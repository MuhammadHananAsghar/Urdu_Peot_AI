#**Collecting Data**

```
from bs4 import BeautifulSoup
import requests
```

```
URL = "http://www.shayaribazar.com/"
html = requests.get(URL).content
soup = BeautifulSoup(html, "html.parser")
```

```
URLS = []
for li in soup.find("div", {"id":"demo1"}).findAll("li"):
  anchor = li.find("a").get("href")
  URLS.append(anchor)
```

```
data = []
for url in URLS:
  try:
    soup = BeautifulSoup(requests.get(url).content, "html.parser")
    link = soup.findAll("ul", class_="pagination")[0].findAll("li")[-2].find("a").get("href")
    data.append([link.split("=")[0], link.split("=")[-1]])
  except:
    pass
```

```
file = open("data.txt", "w")
for i in data:
  for j in range(1, int(i[-1])+1):
    print("Scrapping --> ", i[0]+"="+str(j))
    soup = BeautifulSoup(requests.get(i[0]+"="+str(j)).content, "html.parser")
    for p in soup.findAll("div", class_="main_body_1"):
      file.write(p.find("p").text)
      file.write("\n")
file.close()
```

#**Preprocessing Data**

```
file = open('data.txt', 'r')
readit = file.readlines()
```

```
data = []
for i in readit[:2500]:
  data.append(i)
```

```
print("Number of Lines in Data:",len(data))
```

```
data = '.'.join(data)
```

```
data
```

```
data = data.replace('\ufeff','').replace('\n','').replace('\r','')
```

```
data
```

```
import string
translator = str.maketrans(string.punctuation, ' '*len(string.punctuation)) #map punctuation to space
new_data = data.translate(translator)
```

```
new_data
```

```
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
import pickle
```

```
!rm -rf '/content/tokenized_data.pkl'
```

```
tokenizer = Tokenizer()
tokenizer.fit_on_texts([new_data])

# saving the tokenizer for predict function.
pickle.dump(tokenizer, open('tokenized_data.pkl', 'wb'))
```

```
sequence_data = tokenizer.texts_to_sequences([new_data])[0]
vocab_size = len(tokenizer.word_index) + 1
```

```
print("Sequences Size:",len(sequence_data))
print("Vocabulary Size:",vocab_size)
```

```
sequences = []
for i in range(1, len(sequence_data)):
  words = sequence_data[i - 1 : i + 1]
  sequences.append(words)
```

```
print("Total Sequences:",len(sequences))
```

```
import numpy as np
sequences = np.array(sequences)
```

```
X, Y = sequences[:, 0], sequences[:, 1]
```

```
X = np.array(X)
Y = np.array(Y)
```

```
from tensorflow.keras.utils import to_categorical
```

```
X[:10]
```

```
Y[:10]
```

```
Y = to_categorical(Y, num_classes=vocab_size)
```

```
Y[:10]
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
```

```
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=1))
model.add(LSTM(1000, return_sequences=True))
model.add(LSTM(1000, return_sequences=True))
model.add(LSTM(512))
model.add(Dense(1000, activation="relu"))
model.add(Dense(vocab_size, activation="softmax"))
```

```
model.summary()
```

```
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import TensorBoard

checkpoint = ModelCheckpoint("poet_second.h5", monitor='loss', verbose=1,
    save_best_only=True, mode='auto')

reduce = ReduceLROnPlateau(monitor='loss', factor=0.2, patience=3, min_lr=0.0001, verbose = 1)

logdir='logsnextword1'
tensorboard_Visualization = TensorBoard(log_dir=logdir)
```

```
from tensorflow.keras.optimizers import Adam
model.compile(loss="categorical_crossentropy", optimizer=Adam(lr=0.001), metrics=['accuracy'])
```

```
print("Shape of Input Data:", X.shape)
print("Shape of Output Data:",Y.shape)
```

```
history = model.fit(X, Y, epochs=500, batch_size=200, callbacks=[checkpoint, reduce, tensorboard_Visualization])
```

```
from tensorflow.keras.models import load_model
cnn = load_model('poet.h5')
```

```
def generate_seq(model, tokenizer, enter_text, n_pred):
  in_text, result = enter_text, enter_text
  for _ in range(n_pred):
    encoded = tokenizer.texts_to_sequences([in_text])[0]
    encoded = np.array(encoded)
    yhat = model.predict_classes(encoded)
    out_word = ''
    for word, index in tokenizer.word_index.items():
      if index == yhat:
        out_word = word
        break
    in_text, result = out_word, result + ' '+ out_word
  return result
```
